# Reading txt files using pandas

The two txt files to read are:

    1. Flores_County_NO2.txt
    2. Karen_City_NO2.txt
In both of these files, the first 10 lines contain data unnecessary for this analysis. You can proceed to open these files with notepad for confirmation. 

So, while reading the text file with panda's read_csv method, we will pass value 10 to a keyword argument <b>skiprows</b>.
It will ensure that pandas will skip first 10 lines of the file. 
Additionally, passing <b>","</b> as the value to another keyword argument <b> delimeter</b> will help pandas to seperate values 
and columns with respect to comma.

```python
# reading first file
df_first = pd.read_csv('Flores_County_NO2.txt',skiprows = 10, delimiter = ',')
df_second = pd.read_csv('Karen_City_NO2.txt', skiprows = 10, delimiter = ',')
```

```python
# columns that are completely filled with null values have no use for us. Therefore, its best we remove them
df_first = df_first.dropna(axis = 1)
df_second = df_second.dropna(axis = 1)
```

# Plotting raw data for both dataframes

```python
# lets convert the Date column's values to appropriate format, so that those values can be converted to datetime format


'''
Lets say Date column ,in both dataframes, has a value 20150107 with integer data type.
We will convert that int value to string first.
Then, we will slice that string such that 
if x be the string 20150107,
    x = 20150107
    x[:4] = 2015, as characters in that string in index positions 0 1 2 3 are selected
    x[4:6] = 01
    x[6:] = 07
All of these  values are contatenated to each other using -
Therefore, the final operation is performed such that 20150107 becomes 2015-01-07.
'''

# this will convert the integer values in Date column of both dataframes to string
df_first['Date'] = df_first['Date'].map(str)
df_second['Date'] = df_second['Date'].map(str)

# converting to year-month-day: 2020-02-02
df_first['Date'] = df_first['Date'].apply(lambda x : x[:4] +'-'+ x[4:6] + '-'+x[6:])
df_second['Date'] = df_second['Date'].apply(lambda x : x[:4] +'-'+ x[4:6] + '-'+x[6:])

# converting to datetime format
df_first['Date'] = pd.to_datetime(df_first['Date'])
df_second['Date'] = pd.to_datetime(df_second['Date'])
```

```python
sns.lineplot(x = 'Date', y ='Value', data = df_first)
```

```python
sns.lineplot(x = 'Date', y ='Value', data = df_second)
```

# Plotting daily average 

```python
'''
Using groupby function, we will calculate daily averages for both dataframe using Value column.
'''
df1_avg_day = df_first.groupby('Date')['Value'].mean().reset_index()
df2_avg_day = df_second.groupby('Date')['Value'].mean().reset_index()
```

```python
df1_avg_day
```

```python
df2_avg_day
```

```python
sns.lineplot(x = 'Date', y ='Value', data = df1_avg_day)
sns.lineplot(x = 'Date', y ='Value', data = df2_avg_day)
plt.title('Daily average')
plt.legend(['Flores County', 'Karen City'])
```

```python
# to_period('M') method converts a date in format year-month-day to year-month
# Eg. 2020-01-03 > 2020-01

# Then, converting the updated date to timestamp
df_first['Date_M'] = df_first.Date.dt.to_period('M')
df_first['Date_M'] = df_first['Date_M'].apply(lambda x : x.to_timestamp())

df_second['Date_M'] = df_second.Date.dt.to_period('M')
df_second['Date_M'] = df_second['Date_M'].apply(lambda x : x.to_timestamp())
```

```python
# The above operation can also be performed using lambda
# in our date column, each value is in year-month-day format in datetime datatype
# we can convert the value to string, and take first 7 characters of that string
# eg: x = 2020-08-08 
#      x = str(x)
#      x[:7] = '2020-08-08'[:7] = '2020-08'   
# the resultant value will contain day 1 for every month like 2020-08-01 for all data for 2020 month 8
#  This value can be converted to datetime format using pd.to_datetime() method

df_first['lambda_monthly_experiment'] =  df_first['Date'].apply(lambda x: str(x)[:7])
df_first['lambda_monthly_experiment'] = pd.to_datetime(df_first['lambda_monthly_experiment'])
```

```python
df_first
```

```python
# calculating monthly average using groupby

df1_avg_month = df_first.groupby('Date_M')['Value'].mean().reset_index()
df2_avg_month = df_second.groupby('Date_M')['Value'].mean().reset_index()
```

```python
df2_avg_month
```

```python
sns.lineplot(x = 'Date_M', y ='Value', data = df1_avg_month)
sns.lineplot(x = 'Date_M', y ='Value', data = df2_avg_month)
plt.title('Monthly average')
plt.legend(['Flores County', 'Karen City'])
plt.xlabel('Date')
```
