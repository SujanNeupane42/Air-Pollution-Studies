```python
step = datetime.timedelta(days=28)

result = []
while (dt < end):
    a=int(dt.strftime('%Y%m'))
    result.append(a)
    dt += step
    date=np.unique(result)
    
year=np.array((date/100),dtype=np.int64)
mon=np.array((date)%100,dtype=np.int64)
ts1=year+mon/12

year=np.unique(year)
mon=np.unique(mon)
```

```python
# reading each .nc file
allfile=os.listdir(path)
file=[]

for f in allfile:
    if f.endswith(".nc"):
        file.append(f)
    
file=np.sort(file)
total_months=np.size(file)
total_year=len(year)
```

```python
from tqdm import tqdm
```

```python
# creating a 3 dim numpy array 
NO2=np.zeros((total_months,1800,3600),dtype=np.float32)

print("Reading No2 values for all possible coordinates for all year and months")

# iterating through each month, and putting 2-dim NO2 array into corresponding monthly index of the above initialized array
for j in tqdm(np.arange(total_months)):
    d = Dataset(os.path.join(path, file[j]), mode='r')
#     break
    lat = np.array(d.variables['LatitudeCenter'][:])
    lon = np.array(d.variables['LongitudeCenter'][:])
    no2 = np.array(d.variables['TroposphericNO2'][:])
    NO2[j,:,:]=no2
    

NO2[NO2<=-1.2676506e+30]=np.nan
NO2*=norm_factor #converted to smaller value, norm_factor can be changed in above variable assignment part


#holds yearly data in month level, there should be all 12 months data for all the participating year to work properly
NO2_split_years=np.array_split(NO2, total_year) 
NO2_avg_years=np.zeros((total_year,1800,3600),dtype=np.float32) #holds yearly avg data

print("Calculating Mean for all possible coordinates for each year")
for i in tqdm(np.arange(total_year)):
    NO2_avg_years[i,:,:]=np.nanmean(NO2_split_years[i],axis=0)
```

```python
# calculating levels based on the min and max of avg values for year 2005 in selected region (place/country/state)
no2_first_year=NO2_avg_years[1] # take mean values for first year 2005

#find out the selected region and slice data for that region
ind_lat=np.where((lat>=cur_bottom_lat) & (lat<=cur_top_lat));
ind_lon=np.where((lon>=cur_left_lon) & (lon<=cur_right_lon))

# lat_cur=lat[ind_lat]
# lon_cur=lon[ind_lon]

# this contains the mean data for selected region only
no2_first_year_cur_area=np.array(no2_first_year[ind_lat[0][0]:ind_lat[0][-1]+1,ind_lon[0][0]:ind_lon[0][-1]+1])

#min. mean value in 2005 in selected region
no2_first_year_cur_area_min=int(np.nanmin(no2_first_year_cur_area))

#max. mean value in 2005 in selected region
no2_first_year_cur_area_max=int(np.nanmax(no2_first_year_cur_area))+0.1 
interval=round((no2_first_year_cur_area_max-no2_first_year_cur_area_min)/6.0, 1)

# levels of values in the map, need to update based on the values in particular area/country/state
lvs=np.arange(no2_first_year_cur_area_min,no2_first_year_cur_area_max,interval) 

ln,lt=np.meshgrid(lon,lat)       
```
